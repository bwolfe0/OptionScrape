# SPY Volatility Forecasting with Random Forests and a Trading Strategy Application

## Abstract

##### In finance, the variability of asset prices are uncertain, and understanding their movements is important for investors and financial institutions to manage their risk. As such, these groups are interested in modeling and forecasting asset price volatility. The baseline, and most often used, model to forecast volatility is the linear Heterogeneous Autoregressive (HAR) model. While the HAR performs well given its simplicity, we suspect that more flexible (i.e. machine learning) models can do better. We seek to improve one day ahead volatility forecasts for the S&P 500 (SPY) using machine learning models: SPY is the largest and most traded Exchange Traded Fund in the US and a bellwether for forecasting stock market volatility.  
##### This work will focus on the Random Forest (RF) framework, while a peer examined Neural Networks (NN). The results of the two ML models will be compared. Additionally, we seek to expand the feature set of the RF model beyond the autoregressive lags found in the HAR. Significant improvement was observed when exogenous inputs containing "market information" were added.  
##### Finally, the one-day-ahead forecasts were used to determine buy/sell signals for an option trading strategy based on next the day's volatility: the RF predictions outperformed a peer's Neural Network, the HAR, and a control over the period of 4/1/19 - 8/10/23, with a volatile but impressive cumulative return of over 3000%.

## 1. Introduction

##### Volatility represents how variable the price of an asset is. High volatility results from sharp prices moves during times of turbulence, while low volatility is associated with a more calm market environment. In this project, the terms volatility and variance will be used interchangeably, where volatility is the square root of variance (like standard deviation).
##### The definition of (true) integrated variance is:  
##### $IV_t = \int_{t-1}^{t} \sigma ^2 dt$  
##### However, this quantity is not directly observable, so instead the observable and consistent estimator realiazed variance (RV) is used:  
##### $RV_t = \sum_{i=1}^{M} r_{t,i}^2$
##### which is the sum of intraday log-returns. Our goal is to predict the next day's realized variance, which is typically done with a weighted combination of previous day's RV, where the weights are determined by OLS. The ubiquitous HAR model uses daily, weekly, and monthly lags of RV:
##### $RV_t = \beta_0 + \beta_1*RV_{t-1}^d + \beta_2*RV_{t-1}^w + \beta_3*RV_{t-1}^m$
##### where $RV_t^p = \frac{1}{p}*\sum_{i=1}^{p} RV_{t-i}$ and d=1, w=5, and m=22. The HAR model is popular because of its simplicity and interpretability: each coefficient represents the "importance" of that lag. However, the assumption of a strictly linear relationship is limiting, and we suspect that exogenous variables can incorporate missing information. Additionally, linear models are not well suited for correlated features, which is certainly the case for "lags". Thus, we will examine the more powerful Random Forest framework, which is better suited to handle non-linear relationships, more features, and correlated features.

### **1.3 Rolling Window Model**

##### Market regimes change over long periods of time, generally on the scale of months or years. As a result, it is useful to restrict the model's training data to the most recent history. This is done by refitting the model each day to the previous W days, which allows it to update to the current market regime. The length of the "rolling window" was optimized at 300 for both the Random Forest and HAR from data going back to 2001 (using only the autoregressive lags because of data availability). However, option data was limited (for the IV-based feature), so we chose a 150-day window the RF model to get 150 more obsevations.

#### **Add Realied Quarticity Features**

##### Similar to how Realized Variance is a consistent estimator of Integrated Variance (the second moment of log returns), Realized Quarticity is a consistent estimator for the fourth moment of log returns and is often used in extensions of the HAR. However, we find that the daily, weekly, and monthly RQ lags do not improve model performance. Hence, we will not use them in any further models.

```python
X = pd.concat([X,
               data[['RQ_(t-1)','RQ_t Weekly', 'RQ_t Monthly']]
          ], axis=1)

# RF_6f = RollingWindowRF(X,Y,data['date'],method = 'mse', n_trees=40,w=150)
# dump(RF_6f, 'Data/Trained_Models/RF_6f.sav')
RF_6f = load('Data/Trained_Models/RF_6f.sav')
RF_6f_rel_r2 = round(1 - RF_6f['mse']/(1.86e-8),3)
print(f"RV and RQ Features Only R^2 relatve to HAR: {RF_6f_rel_r2}")
```

```python
X.drop(['RQ_(t-1)','RQ_t Weekly', 'RQ_t Monthly'], axis=1, inplace=True)
X = pd.concat([X,
               data[['r_(t-1)', 'r_t Weekly', 'r_t Monthly']]]
               , axis=1)

# RF_9f = RollingWindowRF(X,Y,data['date'],w=150,n_trees=40, method='mse')
# dump(RF_9f, 'Data/Trained_Models/RF_9f.sav')
RF_9f = load('Data/Trained_Models/RF_9f.sav')
RF_9f_rel_r2 = round(1 - RF_9f['mse']/(1.86e-8),3)
print(f"RV, Return Features R^2 relatve to HAR: {RF_9f_rel_r2}")
```

### **3.4 Feature Importance**

##### For feature importance, we consider SHAP values instead of the popular mean impurity decrease metric for random forests. SHAP values are more powerful since they consider all possible feature combinations and evaluate the direct contribution of each input to the model's output. Importantly, the clever Kernel SHAP method is able to estimate SHAP values without fitting every possible feature permutation, allowing fast computation. Many studies have shown the superiority of SHAP values for "black box" model interpretability. See the widely cited paper *"A Unified Approach to Interpreting Model Predictions"* for more information.

```python
shap_values = RF_best['feature_importance'].T
abs_shap = abs(shap_values)
normalized_shap_values = abs_shap / np.linalg.norm(abs_shap, axis=1, keepdims=True, ord=1)

avg_normalized_shap = np.mean(normalized_shap_values,axis=0)
sorted_indices = sorted(range(len(avg_normalized_shap)), key=lambda i: avg_normalized_shap[i])

renamed_X = X[150:].rename({'avg_ATM_IV': 'Next Day ATM IV', 'ATR': 'Average True Range', 'VIX Close (t-1)': '30D Vix Daily Lag', 
                                               'RV_(t-1)': 'RV Daily Lag', 'r_t Monthly': 'Return Monthly Lag', 'RV_t Monthly': 'RV Monthly Lag',
                                               'r_t Weekly': 'Return Weekly Lag', 'RV_t Weekly': 'RV Weekly Lag', 'r_(t-1)': 'Return Daily Lag',
                                               'pchange 1MT_(t-1)': '1M UST Daily Lag', 'is_friday': 'Friday Indicator'}, axis=1)
sorted_columns = [renamed_X.columns[i] for i in sorted_indices]
sorted_shap = [avg_normalized_shap[i] for i in sorted_indices]

label_colors = {
    'Next Day ATM IV': 'forestgreen',
    'Average True Range': 'forestgreen',
    '30D Vix Daily Lag': 'forestgreen',
    '1M UST Daily Lag': 'forestgreen',
    'Friday Indicator': 'forestgreen',
    'RV Daily Lag': 'darkorange',
    'RV Weekly Lag': 'darkorange',
    'RV Monthly Lag': 'darkorange',
    'Return Daily Lag': 'mediumorchid',
    'Return Weekly Lag': 'mediumorchid',
    'Return Monthly Lag': 'mediumorchid'
}

plt.figure(dpi=125)
plt.barh(sorted_columns, sorted_shap)

plt.xlabel('mean(|SHAP value|) (average impact on model output)', fontsize=14)
plt.title('Feature Importance', fontsize=16)

for i, label in enumerate(plt.gca().get_yticklabels()):
    label_text = label.get_text()
    label.set_color(label_colors.get(label_text, 'black'))

import matplotlib.patches as mpatches
forestgreen_patch = mpatches.Patch(color='forestgreen', label='Exogenous Features')
darkorange_patch = mpatches.Patch(color='darkorange', label='Realized Volatility')
mediumorchid_patch = mpatches.Patch(color='mediumorchid', label='Returns')

plt.legend(handles=[forestgreen_patch, darkorange_patch, mediumorchid_patch], loc='best')

plt.show()
```

##### Up to this point, hyperparameters were chosen to create a flexible Random Forest. Important choices were 40 trees, all features available for each tree, max tree depth of 12, and a minimum of 2 samples required at a terminal node. The number of trees was set arbitrarily at about 25% of the training set size; not too high, to reduce computation time, but large enough to create a robust model. To increase model flexibility, each tree was allowed to use all features, which is not common for a Random Forest. This was done to capture interaction effects between features: if two or more features with important interactions are not often present in the same tree, important relationships could be missed. Thus, the model was similar to bagging, except that each tree trained on a random subset of the data, but not on a random subset of the features. A low value for minimum samples required at a terminal node was chosen to allow to each tree to fit strongly to the training data, given that the inherent randomness of volatility makes it hard to overfit. A moderate value for max depth was chosen to restrict the trees from overfitting. We now examine other values for these parameters to ensure that these choices were reasonable.

##### **Sortino Ratio**

##### The sortino ratio is similar to the popular sharpe ratio, which provides information about an asset or strategy's risk/return tradeoff. While higher returns are generally desirable, it is important to consider the probability and severity of losses. Both ratios are defined as: 
### $\frac{r - r_{rf}}{\sigma}$ 
##### where r is the return, $r_{rf}$ is the risk free rate of return, and $\sigma$ is the standard deviation of returns. They differ in that the sortio ratio only considers "downside" variance: the variance of negative returns. For option trading strategies, and sometimes in general, the downside variance is considered more important.
Note: our trading strategy has such high annualized returns that subtracting the risk free rate does not change the numerator.

```python
rs = [d['Return'] for d in strategy_result_RF]
neg_rs = [d for d in rs if d<0]
sortino_ratio = np.mean(rs) / np.std(neg_rs) * np.sqrt(252)
round(sortino_ratio,3)
```

##### The observed sortino ratio of .774 is below the "minimum" acceptable cutoff of 1, which indicates that the strategy does not have sufficient downside-risk adjusted returns.

##### **Maximum Drawdown**

##### The maximum drawdown of a strategy represents its largest peak-to-trough difference, often on a percentage basis. Unless it is at an all-time high, the cumulative return is below its highest level, so most of the time there is drawdown. The highest drawdown observed is important because if losses become too steep, it is possible that positions are closed and the losses are realized, even if the returns would have eventually recovered.

```python
def calculate_max_drawdown(cumulative_returns):
    peak = cumulative_returns[0]
    trough = cumulative_returns[0]
    max_drawdown = 0.0

    for i in range(1, len(cumulative_returns)):
        if cumulative_returns[i] > peak:
            peak = cumulative_returns[i]
            trough = peak
        elif cumulative_returns[i] < trough:
            trough = cumulative_returns[i]
            drawdown = (peak - trough) / peak
            if drawdown > max_drawdown:
                max_drawdown_date = i
                max_drawdown = drawdown
                max_drawdown_nominal = peak - trough

    return {'max drawdown date': f"{max_drawdown_date}-th day", 'max drawdown percent': round(max_drawdown,3), 'max drawdown nominal': round(max_drawdown_nominal,3)}

calculate_max_drawdown(strategy_return_time_series_RF)
```

##### The maximum drawdown of -80.5% indicates that along the way to 3000% cumulative returns, an investor would have lost 80.5% of their profits. However, this occurred on the 24th day of 941 days, so the nominal losses were not staggering.

##### **Beta**

##### When considering a portfilo or strategy, investors are concerned with how strongly different pieces (and the whole) are correlated with the broader stock market. In a bear market, when most stocks are falling, it is good to have assets that are less correlated with the overall market to minimize losses. Thus, a low correlation with other assets is desirable. In the prominent Capital Asset Pricing Model (CAPM), this correlation is called beta and typically calculated with respect to the S&P 500.
##### $$\beta=\frac{Covariance(r_{strategy}, r_{market})}{Var(r_{market})}$$

##### We calculate our strategy's beta in two ways to ensure correctness. The first is by definition: we simply determine the correlation between the returns of our strategy and of the S&P 500 and normalize it by the market's variance. However, we intuitively expect this to match the slope of a regression line between the returns. If they are uncorrelated, we expect a slope of zero; otherwise the sign will indicate correlation direction. The regression approach also provides a confidence interval for this value, since the true correlation is not observable.

```python
# Calculate beta by definition
round((np.cov(
    pd.Series(strategy_return_time_series_RF).diff().dropna(), 
    (np.exp(data['r_t (return)'][150:].reset_index(drop=True)) -1 )[1:] *100) / np.var((1 - np.exp(data['r_t (return)'][150:].reset_index(drop=True))[1:])*100))[0,1],3)
```

```python
# Calculate beta by regression
import statsmodels.api as sm

X_beta = sm.add_constant((np.exp(data['r_t (return)'][150:].reset_index(drop=True)) - 1)*100) # Adds a constant term to the predictor
model = sm.OLS(pd.Series(strategy_return_time_series_RF).diff().dropna(), X_beta[1:]).fit()

beta = model.params[1]

print(f"Beta: {beta}")
print(model.summary())
```

##### Both methods estimate beta to be -1.72. Additionally, the regression approach provides a 95% confidence interval of [-5.3, 2.2]. Since 0 is in this interval, we accept the null hypothesis that $\beta=0$ and the two are uncorrelated at a 5% significant level. This is an interesting observation since assets or trading strategies with a beta that is zero or negative are very desirable by investors who seek to reduce market risk. With assets or stocks, to achieve a low or negative beta, one must accept lower returns as the price for reduced market risk. However, our strategy has very large returns; the drawback is instead that the returns are highly volatile.

## 6. Conclusions

##### Understanding and predicting stock market volatility are important endeavors for investors and financial institutions to make informed decisions and manage the risk inherent to their practices. Over the last decade (since the paper by Corsi et al.), this task has been primarily accomplishing through use of the Heterogeneous Autoregressive (HAR) model. The HAR model uses a weighted average of the daily, weekly, and monthly "lags" of realized volatility to predict the next day's. While the linear HAR model performs well given its simplicity, we explored the ability a more powerful machine learning framework, specifically Random Forests, to capture the complex and unpredictable dynamics of volatility.  
##### We found that a Random Forest model with the same features as the HAR had a 12% lower one-day-ahead prediction error over the timeframe of 4/1/19 to 8/10/23. Additionally, we examined the ability of exogenous features to improve the Random Forest's predictive ability. The addition of exogenous features led to a 43% error reduction relative to the HAR. Using SHAP values as feature importance, three exogenous features (30-Day VIX, Average True Range, and At-The-Money Implied Volatility) were able to further reduce prediction error by 43% relative to the HAR.
##### The improvement in predictive capabilities was further emphasized through use of an option trading strategy, where the model's prediction determined whether to buy or sell an at-the-money strangle with next-day expiry. Predictions from our best Random Forest model outperformed a peer's Neural Network, the HAR, and a control, with an average daily return of 3.12%. While the volatility of returns from the strategy was undesirable, the strategy had a compelling estimated beta of -1.72.
##### Further work to improve our estimate for at-the-money implied volatility is suggested. One could use more than two contracts, different expiries, or maybe even include more than one implied volatility feature to expand the information available to the model. Additionally, the trading strategy could be improved by including more contracts to cap losses, and implementation of a position sizing strategy is suggested. Furthermore, backtesting the strategy further could provide clarity on the strategy's viability. Specifically, the "volmageddon" episode in 2018 was not examined due to data limitations.
