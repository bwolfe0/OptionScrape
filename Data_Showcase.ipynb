```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Data/Treasury.csv')
HAR_pred = pd.read_csv('Data/predictions_HAR.csv')
RF_pred = pd.read_csv('Data/predictions_RF.csv')
```

```python
dates = data['date'][-180:]

plt.plot(dates,data['RV_t (Realized Variance)'][-180:]**.5,label='Realized Variance')
plt.plot(dates,HAR_pred['values']/np.sqrt(365),label='HAR')
plt.plot(dates,RF_pred['values']/np.sqrt(365),label='RF')
plt.xticks(dates[::30])
plt.legend()
plt.show()
```
